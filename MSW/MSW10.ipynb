10. Řešení obyčejných diferenciálních rovnic

Zadání:
Diferenciální rovnice představují jeden z nejdůležitějších nástrojů každého přírodovědně vzdělaného člověka pro modelování jevů kolem nás. Vaším úkolem je vybrat si nějakou zajímavou soustavu diferenciálních rovnic, která nebyla zmíněna v sešitech z hodin a pomocí vhodné numerické metody je vyřešit. Řešením se rozumí vizualizace jejich průběhu a jiných zajímavých informací, které lze z rovnic odvodit. Proveďte také slovní okomentování toho, co lze z grafu o modelovaném procesu vyčíst.

Lotka-Volterra rovnice jsou následující:
dx/dt = alpha * x - beta * x * y
dy/dt = delta * x * y - gamma * y

kde:

x je počet jedinců kořisti
y je počet jedinců dravce
alpha, beta, delta, gamma jsou parametry ovlivňující růst a interakci mezi populacemi.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Parametry rovnic
alpha = 1.0
beta = 0.5
delta = 0.5
gamma = 2.0

x0 = 2.0  # počáteční počet kořisti
y0 = 1.0  # počáteční počet dravců

# Časový interval a krok
t_start = 0.0
t_end = 10.0
dt = 0.01

# Počet kroků
num_steps = int((t_end - t_start) / dt)

# Inicializace pole pro ukládání hodnot
t_values = np.zeros(num_steps + 1)
x_values = np.zeros(num_steps + 1)
y_values = np.zeros(num_steps + 1)

# Počáteční hodnoty
t_values[0] = t_start
x_values[0] = x0
y_values[0] = y0

# Eulerova metoda pro numerické řešení
for i in range(num_steps):
    t = t_values[i]
    x = x_values[i]
    y = y_values[i]

    # Výpočet derivací
    dx_dt = alpha * x - beta * x * y
    dy_dt = delta * x * y - gamma * y

    # Aktualizace hodnot
    t_values[i + 1] = t + dt
    x_values[i + 1] = x + dt * dx_dt
    y_values[i + 1] = y + dt * dy_dt

# Vizualizace výsledků
plt.plot(t_values, x_values, label='Kořist')
plt.plot(t_values, y_values, label='Dravci')
plt.xlabel('Čas')
plt.ylabel('Počet jedinců')
plt.title('Lotka-Volterra model')
plt.legend()
plt.grid
```
